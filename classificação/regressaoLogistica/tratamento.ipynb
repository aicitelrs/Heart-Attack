```python
# carregar dataset
import numpy as np
import pandas as pd

dados = pd.read_csv('../../data/heart_disease.csv') 

dados.head()
dados.shape
```

### Análise do dataset

```python
# fazer uma copia do dataset, para trabalhar sem riscos
dados2 = pd.DataFrame.copy(dados)
```

```python
# visualizar
dados2.head()
```

```python
# visualizar tipos de dados
dados2.dtypes
```

### Preparar para classificar

```python
# Selecionar colunas relevantes para classificação
previsores = dados2.iloc[:, 1:].values  
previsores
```

```python
#definir meu alvo == target
alvo = dados2.iloc[:, 0].values
alvo
```

```python
dados2.describe() # visualizar para acompanhar progresso
```

### iniciar padronização

```python
from sklearn.preprocessing import StandardScaler # padronização
```

```python
# vai transformar previsores com relação a sua scscala
previsores_esc = StandardScaler().fit_transform(previsores)
previsores_esc
```

```python
# definir dataframe
dados3 = pd.DataFrame(previsores_esc)
dados3
```

```python
dados3.describe() # acompanahr progresso
```

```python
previsores2 = dados3.iloc[:, 1:].values  
previsores2
```

```python
previsores2 = pd.DataFrame(previsores2)
previsores2.head()
```

```python
# vai transformar previsores com relação a sua scscala
previsores2_esc = StandardScaler().fit_transform(previsores2)
previsores2_esc
```

```python
# transformar previsores2 em dataframe
dados4 = pd.DataFrame(previsores2_esc)
dados4
```

```python
dados4.describe() # visualizar
```

#### Salvando
